## **Introduction**

In today’s digital marketplace, understanding consumer purchasing patterns is critical for companies aiming to improve user experiences and drive business growth. Instacart, a major player in online grocery delivery, offers a rich dataset of customer transactions that can be leveraged to predict future buying behavior.

As the online grocery market continues to grow — with sales increasing 31% YoY to $10.3B in February 2025 — competition is intensifying. Major players like Kroger, Aldi, and Walmart are seeing stronger retention rates, putting pressure on Instacart to boost customer loyalty. In this context, understanding and predicting consumer reordering behavior has become more important than ever.

This project focuses on building predictive models to forecast whether a user will reorder a specific product, based on their past shopping history and product characteristics. Through a structured data science pipeline — including data cleaning, feature engineering, sampling strategies, and machine learning model development — we aim to generate actionable insights and optimize prediction performance.

Kaggle dataset:
https://www.kaggle.com/datasets/yasserh/instacart-online-grocery-basket-analysis-dataset

By applying and comparing multiple machine learning algorithms, this study not only highlights the complexities of customer behavior but also showcases the importance of thoughtful feature design and model tuning in real-world predictive analytics.

## Objectives and Goals

- To develop a machine learning-based system that can accurately predict whether a user will reorder a product on Instacart. By analyzing historical purchasing patterns, user behavior, and product information, the model aims to anticipate future reordering actions, which can help in improving recommendation systems and optimizing inventory management.

- To create robust features that effectively capture the interactions between users and products. Special attention is given to the feature engineering process to avoid any form of data leakage.

- To systematically compare and evaluate multiple machine learning models to determine the best-performing method. Specifically, Logistic Regression, Random Forest, XGBoost and Neural networks are implemented and benchmarked.

- To fine-tune the selected models using a thorough hyperparameter search process. Grid Search with cross-validation is employed to explore different combinations of model parameters, optimizing the F1 macro score as the primary evaluation metric. This ensures that model performance is not just satisfactory on training data but also generalizes well to unseen validation data.

## Loading the Data

In this Code, we will be using the 'Instacart Market Basket Analysis'.

```python
import pandas as pd

# Load the raw data - aisles.csv
df_raw_aiseles = pd.read_csv(abspath_curr + 'data/aisles.csv',
                           header=0)
# Load the raw data - departments.csv
df_raw_departments = pd.read_csv(abspath_curr + 'data/departments.csv',
                           header=0)
# Load the raw data - order_products__prior.csv
df_raw_order_products_prior = pd.read_csv(abspath_curr + 'data/order_products__prior.csv',
                           header=0, nrows=1_000_000)
# Load the raw data - order_products__train.csv
df_raw_order_products_train = pd.read_csv(abspath_curr + 'data/order_products__train.csv',
                           header=0)
# Load the raw data - orders.csv
df_raw_orders = pd.read_csv(abspath_curr + 'data/orders.csv',
                           header=0)
# Load the raw data - products.csv
df_raw_products = pd.read_csv(abspath_curr + 'data/products.csv',
                           header=0)
```

## Dataset Merging

In this code, we will be merging the 6 feature datasets namely: aisles, departments, order_products_prior, order_products_train, orders, products.

```python
# Merge Products, Aiseles and departments
df_products_merged = df_raw_products.merge(df_raw_aiseles, on='aisle_id', how='left')
df_products_merged = df_products_merged.merge(df_raw_departments, on='department_id', how='left')
```

```python
# Merge with order_products_prior
df_order_prior_merged = df_products_merged.merge(df_raw_order_products_prior, on='product_id', how='left')

# Merge with order_products_train
df_order_train_merged = df_products_merged.merge(df_raw_order_products_train, on='product_id', how='left')

```

```python
# Split orders based on eval_set
orders_prior = df_raw_orders[df_raw_orders['eval_set'] == 'prior']
orders_train = df_raw_orders[df_raw_orders['eval_set'] == 'train']
orders_test  = df_raw_orders[df_raw_orders['eval_set'] == 'test']
```

```python
# Merge final with orders-prior
df_prior_merged = df_order_prior_merged.merge(orders_prior, on='order_id')

# Merge final with orders-train
df_train_merged = df_order_train_merged.merge(orders_train, on='order_id')

# Merge both prior and train
df_prior_train_merged = pd.concat([df_prior_merged, df_train_merged], ignore_index=True)
```

```python
# Merge with test
df_prior_train_test = df_prior_train_merged.drop(['order_id', 'eval_set', 'order_number', 'order_dow', 'order_hour_of_day', 'days_since_prior_order'], axis=1)
df_test_merged = orders_test.merge(df_prior_train_test, on='user_id', how='left')
```

```python
# There are a lot of null values created in days_since_order, we do not need those rows, lets drop them
# If we do not drop now, we will face issue with  feature engineering
df_test_merged = df_test_merged.dropna()
df_train_merged = df_train_merged.dropna()
df_prior_merged = df_prior_merged.dropna()
```

```python
# Save final datasets
df_prior_merged.to_csv("prior_merged.csv", index=False)
df_train_merged.to_csv("train_merged.csv", index=False)
df_test_merged.to_csv("orders_test.csv", index=False)
```

## Feature Engineering

Feature engineering was performed across eight key dimensions to enhance model performance. New features captured user-product interactions, user behavior, product popularity, cart order tendencies, and aggregated patterns at the aisle, department, and order-time levels, along with overall reorder trends.

##### User Product Features

These capture the relationship between a specific user and a product, such as how often they buy it, when they tend to order it, and where it appears in their cart.

```python
#  The number of times user ordered a product
def add_user_product_cnt(df):
    user_product_cnt = df.groupby(['user_id', 'product_id']).size().reset_index(name='user_product_cnt')
    df = df.merge(user_product_cnt, on=['user_id', 'product_id'], how='left')
    return df

df_train = add_user_product_cnt(df_train)
df_val   = add_user_product_cnt(df_val)
df_test  = add_user_product_cnt(df_test)

df_train.head()
```

```python
# Total orders for each user
def add_user_total_product_cnt(df):
    user_total_product_cnt = df.groupby('user_id').size().reset_index(name='user_total_product_cnt')
    df = df.merge(user_total_product_cnt, on='user_id', how='left')
    return df

df_train = add_user_total_product_cnt(df_train)
df_val   = add_user_total_product_cnt(df_val)
df_test  = add_user_total_product_cnt(df_test)
df_val.head()
```

```python
# How many times on an average a user orders a product in a day in a week
def add_user_product_dow_mean(df):
    user_product_dow_mean = df.groupby(['user_id', 'product_id'])['order_dow'].mean().reset_index(name='user_product_dow_mean')
    df = df.merge(user_product_dow_mean, on=['user_id', 'product_id'], how='left')
    return df

df_train = add_user_product_dow_mean(df_train)
df_val   = add_user_product_dow_mean(df_val)
df_test  = add_user_product_dow_mean(df_test)
```

```python
# Which hour of the day does the user most frequently order a product?
def add_user_product_hod_mode(df):
    user_product_hod_mode = (
        df
        .groupby(['user_id', 'product_id'])['order_hour_of_day']
        .agg(lambda x: x.mode().iloc[0])  # In case of tie, picks first mode
        .reset_index(name='user_product_hod_mode')
    )
    df = df.merge(user_product_hod_mode, on=['user_id', 'product_id'], how='left')
    return df

df_train = add_user_product_hod_mode(df_train)
df_val   = add_user_product_hod_mode(df_val)
df_test  = add_user_product_hod_mode(df_test)

df_train.head()
```

```python
# Weeknd and Weekday
def add_weekend_flag(df):
    df = df.copy()
    df['Weekend'] = [1 if x in [0, 1] else 0 for x in df['order_dow']]
    return df

df_train = add_weekend_flag(df_train)
df_val   = add_weekend_flag(df_val)
df_test  = add_weekend_flag(df_test)
```

```python
# How many orders would a user place if weekday or weekend on average
import numpy as np

def add_user_product_daytype_avg(df):
    df = df.copy()
    df['Weekend'] = df['order_dow'].isin([0, 1]).astype(int)

    user_product_wknd_count = df.groupby(['user_id', 'product_id', 'Weekend']).size().reset_index(name='orders_by_daytype')

    # Normalize by number of weekend/weekday days in a week
    user_product_wknd_count['orders_by_daytype_avg'] = np.where(
        user_product_wknd_count['Weekend'] == 1,
        user_product_wknd_count['orders_by_daytype'] / 2,
        user_product_wknd_count['orders_by_daytype'] / 5
    )

    user_product_daytype_avg = user_product_wknd_count.drop(columns=['orders_by_daytype', 'Weekend'])

    # Sum both weekday and weekend averages (so we get one row per user-product)
    user_product_daytype_avg = user_product_daytype_avg.groupby(['user_id', 'product_id'])['orders_by_daytype_avg'].sum().reset_index()

    df = df.merge(user_product_daytype_avg, on=['user_id', 'product_id'], how='left')
    return df

df_train = add_user_product_daytype_avg(df_train)
df_val   = add_user_product_daytype_avg(df_val)
df_test  = add_user_product_daytype_avg(df_test)

df_train.head()
```

```python
# Min,max and mean of the position of the product in the cart
def add_user_product_atc_stats(df):
    # Mean
    user_product_atc_mean = df.groupby(['user_id', 'product_id'])['add_to_cart_order'].mean().reset_index()
    user_product_atc_mean.columns = ['user_id', 'product_id', 'user_product_atc_mean']

    # Min
    user_product_atc_min = df.groupby(['user_id', 'product_id'])['add_to_cart_order'].min().reset_index()
    user_product_atc_min.columns = ['user_id', 'product_id', 'user_product_atc_min']

    # Max
    user_product_atc_max = df.groupby(['user_id', 'product_id'])['add_to_cart_order'].max().reset_index()
    user_product_atc_max.columns = ['user_id', 'product_id', 'user_product_atc_max']

    # Merge into main df
    df = df.merge(user_product_atc_mean, on=['user_id', 'product_id'], how='left')
    df = df.merge(user_product_atc_min, on=['user_id', 'product_id'], how='left')
    df = df.merge(user_product_atc_max, on=['user_id', 'product_id'], how='left')

    return df

df_train = add_user_product_atc_stats(df_train)
df_val   = add_user_product_atc_stats(df_val)
df_test  = add_user_product_atc_stats(df_test)

df_val.head()
```

```python
# Total number of orders for each user

def add_user_total_orders(df):
    if 'orders_total' in df.columns:
        df = df.drop(columns='orders_total')

    user_order_sum = df.groupby('user_id')['order_id'].count().reset_index(name='orders_total')
    df = df.merge(user_order_sum, on='user_id', how='left')

    return df

df_train = add_user_total_orders(df_train)
df_val   = add_user_total_orders(df_val)
df_test  = add_user_total_orders(df_test)
```

```python
# Number of orders reordered
def add_user_reorder_stats(df):
    user_sum_reord = df.groupby('user_id').agg(user_sum_reord=('reordered', 'sum')).reset_index()
    df = df.merge(user_sum_reord, on='user_id', how='left')
    return df

df_train = add_user_reorder_stats(df_train)
df_val   = add_user_reorder_stats(df_val)
df_test  = add_user_reorder_stats(df_test)
```

```python
# Proportion of orders reordered by users
# TRAIN
df_train['user_reord_prop'] = df_train['user_sum_reord'] / df_train['orders_total']

# VAL
df_val['user_reord_prop'] = df_val['user_sum_reord'] / df_val['orders_total']

# TEST
df_test['user_reord_prop'] = df_test['user_sum_reord'] / df_test['orders_total']
```

```python
# Proportion of products that are atleast ordered once by the user

# TRAIN
user_prod_reord_prop_train = df_train.groupby(['user_id', 'product_id']).agg({'reordered': sum}).reset_index()
user_prod_reord_prop_train['reordered'] = user_prod_reord_prop_train['reordered'].apply(lambda x: 1 if x > 0 else 0)
user_prod_reord_prop_train = user_prod_reord_prop_train.groupby(['user_id']).agg({'product_id': 'count', 'reordered': sum}).reset_index()
user_prod_reord_prop_train['user_prod_reord_prop'] = user_prod_reord_prop_train['reordered'] / user_prod_reord_prop_train['product_id']
user_prod_reord_prop_train.drop(['product_id', 'reordered'], axis=1, inplace=True)

df_train = df_train.merge(user_prod_reord_prop_train, on='user_id', how='left')

df_train['user_reord_prop'] = df_train['user_sum_reord'] / df_train['orders_total']

# VAL
user_prod_reord_prop_val = df_val.groupby(['user_id', 'product_id']).agg({'reordered': sum}).reset_index()
user_prod_reord_prop_val['reordered'] = user_prod_reord_prop_val['reordered'].apply(lambda x: 1 if x > 0 else 0)
user_prod_reord_prop_val = user_prod_reord_prop_val.groupby(['user_id']).agg({'product_id': 'count', 'reordered': sum}).reset_index()
user_prod_reord_prop_val['user_prod_reord_prop'] = user_prod_reord_prop_val['reordered'] / user_prod_reord_prop_val['product_id']
user_prod_reord_prop_val.drop(['product_id', 'reordered'], axis=1, inplace=True)

df_val = df_val.merge(user_prod_reord_prop_val, on='user_id', how='left')

df_val['user_reord_prop'] = df_val['user_sum_reord'] / df_val['orders_total']

# TEST
user_prod_reord_prop_test = df_test.groupby(['user_id', 'product_id']).agg({'reordered': sum}).reset_index()
user_prod_reord_prop_test['reordered'] = user_prod_reord_prop_test['reordered'].apply(lambda x: 1 if x > 0 else 0)
user_prod_reord_prop_test = user_prod_reord_prop_test.groupby(['user_id']).agg({'product_id': 'count', 'reordered': sum}).reset_index()
user_prod_reord_prop_test['user_prod_reord_prop'] = user_prod_reord_prop_test['reordered'] / user_prod_reord_prop_test['product_id']
user_prod_reord_prop_test.drop(['product_id', 'reordered'], axis=1, inplace=True)

df_test = df_test.merge(user_prod_reord_prop_test, on='user_id', how='left')

df_test['user_reord_prop'] = df_test['user_sum_reord'] / df_test['orders_total']
```

### Creating the pipeline of models

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier

random_seed = 42


models = {'lr': LogisticRegression(class_weight='balanced', random_state=random_seed),
           'rf': RandomForestClassifier(random_state=random_seed),
           'xgb': XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=random_seed)
          }
```

```python
from sklearn.pipeline import Pipeline

pipes = {}

for acronym, model in models.items():
    pipes[acronym] = Pipeline([('model', model)])
```

### GridSearchCV

```python
param_grids = {}
```

Note: The below section of the code can take more than 30 minutes to run.

```python
# The parameter grid of tol
tol_grid = [10 ** -5, 10 ** -4, 10 ** -3]

# The parameter grid of C
C_grid = [1, 10]

# The number of model estimators
model_n_estimators = [50, 100, 200]

# Model max depth-rf
model_max_depth1 = [10, 20]

# Model max depth-xgb
model_max_depth2 = [3, 5]

# Model learning rate
model_learning_rate = [0.001, 0.1, 0.2]



# Update param_grids
param_grids['lr'] = [{'model__tol': tol_grid,
                     'model__C': C_grid}]

param_grids['rf'] = [{'model__max_depth': model_max_depth1,
                      'model__n_estimators': model_n_estimators}]

param_grids['xgb'] = [{'model__max_depth': model_max_depth2,
                       'model__n_estimators': model_n_estimators,
                       'model__learning_rate': model_learning_rate}]

```

#### Creating the directory for the cv results produced by GridSearchCV

```python
# Make directory
directory = os.path.dirname(abspath_curr + '/result/mnist/cv_results/GridSearchCV/')
if not os.path.exists(directory):
    os.makedirs(directory)
```

#### Tuning the hyperparameters

```python
from sklearn.model_selection import GridSearchCV

# The list of [best_score_, best_params_, best_estimator_] obtained by GridSearchCV
best_score_params_estimator_gs = []

# For each model
for acronym in pipes.keys():
    # GridSearchCV
    gs = GridSearchCV(estimator=pipes[acronym],
                      param_grid=param_grids[acronym],
                      scoring='f1_macro',
                      n_jobs=2,
                      cv=ps,
                      return_train_score=True)

    # Fit the pipeline
    gs = gs.fit(X_train_val, y_train_val)

    # Update best_score_params_estimator_gs
    best_score_params_estimator_gs.append([gs.best_score_, gs.best_params_, gs.best_estimator_])

    # Sort cv_results in ascending order of 'rank_test_score' and 'std_test_score'
    cv_results = pd.DataFrame.from_dict(gs.cv_results_).sort_values(by=['rank_test_score', 'std_test_score'])

    # Get the important columns in cv_results
    important_columns = ['rank_test_score',
                         'mean_test_score',
                         'std_test_score',
                         'mean_train_score',
                         'std_train_score',
                         'mean_fit_time',
                         'std_fit_time',
                         'mean_score_time',
                         'std_score_time']

    # Move the important columns ahead
    cv_results = cv_results[important_columns + sorted(list(set(cv_results.columns) - set(important_columns)))]

    # Write cv_results file
    cv_results.to_csv(path_or_buf=abspath_curr + '/result/mnist/cv_results/GridSearchCV/' + acronym + '.csv', index=False)

# Sort best_score_params_estimator_gs in descending order of the best_score_
best_score_params_estimator_gs = sorted(best_score_params_estimator_gs, key=lambda x : x[0], reverse=True)

# Print best_score_params_estimator_gs
pd.DataFrame(best_score_params_estimator_gs, columns=['best_score', 'best_param', 'best_estimator'])
```

#### Experiment Results

After performing Grid Search with Cross-Validation, we identified the best models and hyperparameters based on the validation F1 Macro Score.

As you can see from the table:

XGBoost Classifier delivered the best performance with a score of 0.9022. Its best parameters included a learning rate of 0.2 and a max depth of 3. This configuration allowed the model to efficiently capture complex user-product reorder patterns without overfitting.

The Decision Tree Classifier followed closely, achieving a score of 0.9011 with a max depth of 20 and 200 estimators. Although it performed well, it was slightly less powerful than XGBoost in modeling subtle behaviors.

Logistic Regression achieved a score of 0.8919. It used C=1 and a tolerance of 0.001. While Logistic Regression offered fast training and interpretability, it was slightly less effective at capturing the complexity of reordering patterns compared to ensemble methods.

Overall, the results confirmed that ensemble-based models like XGBoost and Decision Trees provided stronger predictive performance on our large and feature-rich dataset.
